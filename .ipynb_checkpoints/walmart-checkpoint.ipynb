```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
walmart=pd.read_csv("Walmart_Sales.csv")

walmartv2=walmart.copy()
```

```python
walmartv2.groupby("Date")["Weekly_Sales"].sum().reset_index()
walmartv2['Weekly_Sales'].astype(int)
```

```python
walmartv2['Date']=pd.to_datetime(walmartv2['Date'], format='%d-%m-%Y')
sales_trend=walmartv2.groupby(walmartv2["Date"].dt.to_period("M"))["Weekly_Sales"].sum().round().reset_index()

plt.figure(figsize=[25,9])
plt.plot(sales_trend["Date"].astype(str),sales_trend['Weekly_Sales'],"b.-")

plt.show()
```
