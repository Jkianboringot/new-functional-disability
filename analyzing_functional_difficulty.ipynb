```python
import pandas as pd 
import os
import numpy as np
import warnings
warnings.filterwarnings("ignore")
pwd=os.getcwd()
dataset=pd.read_csv(r"clean_functional_difficulty.csv")
region_total_population=pd.read_csv("region_total_population.csv")
```

```python
dataset.drop(columns='Unnamed: 0',inplace=True)
df=dataset

df.rename(columns={"Household Population 5 Years Old and Over with Functional Difficulty":"Total Functional Difficulty"},inplace=True)
```

```python
df["Status"].unique()
```

```python
df["Region"] = df["Region"].str.upper().str.strip()
df["Status"] = df["Status"].str.upper()
df["Disability"] = df["Disability"].str.upper()
df["Sex"] = df["Sex"].str.upper()
df["Province"] = df["Province"].str.upper().str.strip()
df
```

```python
region_age_total=df.groupby("Region",as_index=False)[['Childern Age: 5-14', 'Adolescents: 15-19', 'Young adult: Age 20 - 34',
       'Adults: 35–54 years', 'Senior: over 55 years']].sum()
region_age_total
```

```python
region_total_population["Region"]=region_total_population["Region"].str.upper().str.strip()
```

```python
disability_by_region = (
    df.loc[(df["Sex"] == "ALL SEX")& (df["Status"] != "ALL") ]
    .groupby(["Disability", "Region","Status"], as_index=False)["Total Functional Difficulty"]
    .sum()
    .assign(
        Region=lambda x: x["Region"]
        
        .str.replace("^NATIONAL CAPITAL REGION$", "NATIONAL CAPITAL REGION (NCR)", regex=True)
    )
)
```

```python

# i need to unfilter this by severe
# i need  to standardize the region names and status 
total_disability_population=pd.merge(disability_by_region,region_total_population,on="Region",how="left")
total_disability_population["Person affected by disability per 100k"]=np.ceil(total_disability_population['Total Functional Difficulty']
                                                                       /total_disability_population['Population']
                                                                       * 100000)

def func(x):
    return total_disability_population.loc[total_disability_population["Status"]==str(x).upper().strip()]
func("mild") # will put in try catch later to catch errors like spelling and spacing as whre it can filter to certain disability
```
