```python
import pandas as pd 
import os
import numpy as np
import warnings
warnings.filterwarnings("ignore")
pwd=os.getcwd()
dataset=pd.read_csv(pwd + "//clean_functional_difficulty.csv")
dataset.drop(columns=["PSGC",'Unnamed: 0'])
region_total_population=pd.read_csv("region_total_population.csv")
```

```python

df=dataset
df.head()
df["Region"]=df["Region"].str.strip()
df.rename(columns={"Household Population 5 Years Old and Over with Functional Difficulty":"5 Years and Over with Functional Difficulty"},inplace=True)
```

```python
df
```

```python
region_total_population["Region"]=region_total_population["Region"].str.upper()
```

```python
disability_by_region = (
    df.loc[(df["Sex"] == "All Sex") & (df["Status"] == "Severe")]
    .groupby(["Disability", "Region"], as_index=False)["5 Years and Over with Functional Difficulty"]
    .sum()
    .assign(
        Region=lambda x: x["Region"]
        .str.upper()
        .str.replace("^NATIONAL CAPITAL REGION$", "NATIONAL CAPITAL REGION (NCR)", regex=True)
    )
)
```

```python
total_severe_disability_population=pd.merge(disability_by_region,region_total_population,on="Region",how="left")




total_disability_population=pd.merge(disability_by_region,region_total_population,on="Region",how="left")
total_disability_population["Person affected by disability per 100k"]=(total_disability_population['5 Years and Over with Functional Difficulty']
                                                                       /total_disability_population['Population']
                                                                       * 100000)
total_severe_disability_population
```

```python
most_affected_age=(df.loc[(df["Sex"] == "All Sex") & (df["Status"] == "All")]
    .groupby(["Disability","Region"], as_index=False)[['Childern Age: 5-14', 'Adolescents: 15-19', 'Young adult: Age 20 - 34',
       'Adults: 35â€“54 years', 'Senior: over 55 years']]
    .sum())

MAA_melted=most_affected_age.melt(id_vars=["Region","Disability"],var_name="Age Category",value_name="value")
idx=MAA_melted.groupby(["Region","Disability"])["value"].idxmax()
MAA_melted_filtered=MAA_melted.loc[idx,["Region","Disability","value","Age Category"]]
```

```python


MAA_melted_filtered.pivot_table(index="Region",columns="Disability",
                       values=["Age Category","value"],
                       aggfunc={"Age Category":"max","value":"max"})
```

```python
gender_severety=(df.loc[(df["Sex"] != 'All Sex') & (df["Status"] == 'Severe')]
                 .groupby(["Region","Sex","Disability"],as_index=False)
                 ["5 Years and Over with Functional Difficulty"]
                 .sum().assign(Region=lambda x:x["Region"].str.upper()
                .str.replace("^NATIONAL CAPITAL REGION$", "NATIONAL CAPITAL REGION (NCR)", regex=True)))



gender_serverity_disability=pd.merge(gender_severety,region_total_population,on="Region",how="left").assign(population=lambda x:x["Population"]/2)
gender_serverity_disability["Gender affected by disability per 100k"]=gender_serverity_disability["5 Years and Over with Functional Difficulty"]/(gender_serverity_disability['Population']/2)*100000
gender_serverity_disability.rename(columns={"Population":"Total Population"},inplace=True)
gender_serverity_disability.rename(columns={"population":"50/50 Population"},inplace=True)
```
